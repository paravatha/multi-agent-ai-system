```python
import os

from dotenv import load_dotenv

load_dotenv()

mlflow_tracking_uri = os.getenv("MLFLOW_TRACKING_URI")
gemini_model_name = os.getenv("GEMINI_MODEL_NAME")
print("MLflow Tracking URI:", mlflow_tracking_uri)
print("Gemini Model Name:", gemini_model_name)
```

```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search

print("✅ ADK components imported successfully.")
```

```python
google_search_agent = Agent(
    name="helpful_assistant",
    model="gemini-2.5-flash-lite",
    description="A simple agent that can answer general questions.",
    instruction="""You are a helpful assistant.
      Use Google Search for current info or if unsure.""",
    tools=[google_search],
)

print("✅ google_search_agent defined.")
```

```python
runner = InMemoryRunner(agent=google_search_agent)

print("✅ Runner created.")
```

```python
response = await runner.run_debug(
    "What is Agent Development Kit from Google? What languages is the SDK available in?"
)
print("✅ Runner execution completed.")
```

```python
response = await runner.run_debug("What's the weather in Washington DC?")
print("✅ Runner execution completed.")
```
